```python
import numpy as np
N = 10**4
winsConservative = 0
winsSwitcher = 0
wins = 0
for i in range(N):
    doors = np.array(['car', 'goat', 'goat'])
    np.random.shuffle(doors)
    door_chosen = np.random.randint(0, 3)
    choice = doors[door_chosen]
    
    
    #print(choice)

    while True:
        opened = np.random.randint(0, 3)
        #print('---', doors[opened])
        if doors[opened] == 'goat' and opened != door_chosen: 
            break

    #print(np.delete(doors, opened), choice)
    
    
    if choice == 'car': winsConservative += 1 
    if np.delete(doors, [door_chosen, opened]) == 'car': winsSwitcher += 1
    if np.random.choice(np.delete(doors, opened)) == 'car': wins += 1

        
print(winsConservative, winsSwitcher, wins)
```
