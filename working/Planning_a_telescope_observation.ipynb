```python
import numpy as np
import pylab as plt

def weather_today(weather, A):

    if weather == 'cloudy':
        if A < 0.5:
            weather = 'cloudy'
        else: 
            weather = 'clear'
        

    if weather == 'clear':
        if A < 0.1:
            weather = 'cloudy'
        else: 
            weather = 'clear'
            
    return(weather)

N = 100000      

cumsum_clear = []
weather = 'cloudy'    
clear = 0
for i in range(1, N):
    weather = weather_today(weather, np.random.rand())
    #print(weather)
    if weather == 'clear':
        
        clear += 1
        cumsum_clear.append(clear/i)
        #print(clear)

plt.figure(figsize = (14, 8))
#plt.plot(np.linspace(1, N, np.array(cumsum_clear).size), cumsum_clear)

n_burnin = 20000
cumsum_clear = cumsum_clear[n_burnin:-1]

plt.plot(np.linspace(1, N-n_burnin, np.array(cumsum_clear).size), cumsum_clear)
```
